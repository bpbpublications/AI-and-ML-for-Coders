```python
# import Numpy, Pandas libraries
import numpy as np
import pandas as pd
```

```python
# download the Zip file from
# the UCI repo : https://archive.ics.uci.edu/static/public/602/dry+bean+dataset.zip
# to the current working directory
import urllib.request
url = 'https://archive.ics.uci.edu/static/public/602/dry+bean+dataset.zip'
urllib.request.urlretrieve(url, 'dry+bean+dataset.zip')
```

```python
# Save zip file https://archive.ics.uci.edu/static/public/602/dry+bean+dataset.zip
# to current directory and unzip
import zipfile
with zipfile.ZipFile('dry+bean+dataset.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
# Load the extracted XLSX file
# using Pandas
df = pd.read_excel('/content/DryBeanDataset/Dry_Bean_Dataset.xlsx')
```

```python
# import matplotlib and seaborn library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot the scatter plots of Area / Perimiter columns
# in the dataframe to compare them with one another
# using the matplotlib library
plt.scatter(df['Area'], df['Perimeter'])
plt.show()
```

```python
# generate same scatter plot with title
# axis labels
plt.scatter(df['Area'], df['Perimeter'])
plt.title('Area vs Perimeter')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.show();
```

```python
# create the same scatter plot
# this time using seaborn library
# use title and axis labels
# and show the plot
sns.scatterplot(x='Area', y='Perimeter', data=df)
plt.title('Area vs Perimeter')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.show()
```

```python
# generate scatter plots for the combination
# of all variables in the dataframe, using the CLASS column
# as the color identifier using Seaborn
sns.pairplot(df, hue='Class')
plt.show()
```

```python
# draw a stripplot using the Class column
# from the dataframe
sns.stripplot(x='Class', y='Perimeter', data=df)
plt.show()
```

```python
# perform scaling of the Perimeter column
# using Min Max scaler and draw the same stripplot again
# using the scaled column
# and show the plot
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Perimeter'] = scaler.fit_transform(df[['Perimeter']])
sns.stripplot(x='Class', y='Perimeter', data=df)
plt.show()
```

```python
# Probability distribution function
# of the Area column using the Class column
# as the Hue from the dataframe.
plt.figure(figsize=(14,14));
sns.FacetGrid(df, hue='Class') \
    .map(sns.distplot, 'Area') \
    .add_legend();
plt.show();
```
